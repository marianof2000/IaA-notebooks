```python
'''
Ejercicio 9: Crear tres listas y ordenarlas en forma ascendente utilizando un método para cada lista:
métodos de selección, inserción y burbujeo. ¿Qué cambia para ordenar en forma descendente?
'''
# Ej 09
import random
from timeit import timeit
import time

def ordBurbujeo(lista):
    '''
    El algoritmo de clasificación de intercambio o de la burbuja se basa en el principio de comparar pares de elementos
    adyacentes e intercambiarlos entre sí hasta que estén todos ordenados.
    '''
    longitud = len(lista)
    for j in range(longitud - 1):
        for i in range(longitud - 1):
            if lista[i] > lista[i + 1]:
                # Secuencia de intercambio de valores
                aux = lista[i]
                lista[i] = lista[i + 1]
                lista[i + 1] = aux
    return lista

def ordInsercion(lista):
    '''
    Este método consiste en insertar un elemento en el vector en una parte ya ordenada
    de este vector y comenzar de nuevo con los elementos restantes.
    '''
    longitud = len(lista)
    for j in range(1, longitud):
        aux = lista[j]
        k = j - 1
        cambio = False
        while not cambio and k >= 0:
            if aux < lista[k]:
                lista[k + 1] = lista[k]
                k -= 1
            else:
                cambio = True
        lista[k + 1] = aux
    return lista

def ordSeleccion(lista):
    '''
    Este método se basa en buscar el elemento menor del vector y colocarlo en primera posición.
    Luego se busca el segundo elemento más pequeño y se coloca en la 
    segunda posición, y así sucesivamente.
    '''
    longitud = len(lista)
    for j in range(0, longitud -1):
        aux = lista[j]
        k = j
        for i in range(j + 1, longitud):
            if aux > lista[i]:
                aux = lista[i]
                k = i
        lista[k] = lista[j]
        lista[j] = aux
    return lista
    pass

elementos = 5000

lista1 = creaLista(1, 100, elementos)
print(lista1)
inicio = time.time()
print(f'{ordBurbujeo(lista1)}\n')
fin = time.time()
print()
print(fin - inicio)

lista2 = creaLista(1, 100, elementos)
print(lista2)
inicio = time.time()
print(f'{ordInsercion(lista2)}\n')
fin = time.time()
print()
print(fin - inicio)

lista3 = creaLista(1, 100, elementos)
print(lista3)
inicio = time.time()
print(f'{ordSeleccion(lista3)}\n')
print()
fin = time.time()
print(fin - inicio)
```

```python
'''
Ejercicio 10: Utilizando búsqueda binaria sobre una lista ordenada realizar la búsqueda de valores e
informar si se encuentran o no en la lista, finalizar las búsquedas con -1 e informar cuantas
búsquedas fueron exitosas y en cuantas no se encontró el valor buscado.
'''
# Ej 10
def busquedaBinaria(lista, item):
    # primer índice de la lista
    primero = 0
    # último índice de la lista
    ultimo = len(lista) - 1
    # aún no lo encontró
    encontrado = False
    while primero <= ultimo and not encontrado:
        # ubica la mitad de las lista aunque la longitud sea par o impar
        puntoMedio = (primero + ultimo) // 2
        # Revisa si el item buscado está en el medio de la lista
        if lista[puntoMedio] == item:
            encontrado = True
        else:
            # Revisa si el ítem buscado es menor al del medio de la lista
            if item < lista[puntoMedio]:
                # Toma la mitad de la izquierda ahora como toda la lista
                ultimo = puntoMedio-1
            else:
                # Si el item es mayor al medio de la lista
                # Toma la mitad de la derecha ahora como toda la lista
                primero = puntoMedio+1	
        # Repite todo el proceso hasta que no queda lista para buscar o lo haya encontrado
    # Devuelve si lo encontró o no
    return encontrado

# Usa la funsción del ej 02 para crear la lista	
listaPrueba = creaLista(1, 100, 100)
# Usa la función del ej 09 para ordenar la lista
listaOrdenada = ordBurbujeo(listaPrueba)

print(listaOrdenada)

while True:
    numero = int(input('Ingrese el item a buscar: '))
    if numero != -1:
        if busquedaBinaria(listaOrdenada, numero):
            print(f'El número se encuentra en la lista\n')
        else:
            print(f'El número no se encuentra en la lista\n')
    else:
        print('Adiós')
        break
```
